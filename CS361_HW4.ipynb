```python
import numpy as np
import matplotlib.pyplot as plt
```

### Problem 5

```python
p = np.linspace(0.1,1,10)            # array for p

def trial(p):
    presence = np.random.rand(10)    # present if less than p, absent otherwise
    gender = np.random.rand(10)      # female if less than 0.5, male otherwise
    
    # calculate the numbers of males and females
    m = np.count_nonzero(np.logical_and(presence < p, gender >=0.5))
    f = np.count_nonzero(np.logical_and(presence < p, gender < 0.5))

    return m,f
```

```python
exp_value = np.zeros(10)

for i in range(10):
    temp = 0.0
    count = 0
    prob = p[i]
    for j in range(int(1e5)):       # run 1e5 times
        m,f = trial(prob)
        if f >= 2:                  # only count the number if the aircraft flies
            temp += m+f
            count += 1
    exp_value[i] = temp / count     # calculate the averge
```

```python
plt.plot(p, exp_value, marker='o')
plt.xlabel('value for p')
plt.ylabel('average number of presence')
plt.title('Simulation of the number of passengers if the aircraft flies on difference values of p')
plt.show()
```
